```python
acc_2020 = pd.read_csv('accident2020.csv')
#print(acc_2020.head())

acc_2020_cleaned = acc_2020.dropna()
print(acc_2020_cleaned.head())
print(acc_2020_cleaned['ปีที่เกิดเหตุ'].isnull().values.any())
```

```python
province_2020_count = acc_2020_cleaned['จังหวัด'].value_counts()
province_2020_count_df = pd.DataFrame({'province':province_2020_count.index,'count':province_2020_count.values,'year':2020})
print(province_2020_count_df.head(10))
```

```python
acc_2021 = pd.read_csv('accident2021.csv')
#print(acc_2020.head())

acc_2021_cleaned = acc_2021.dropna()
print(acc_2021_cleaned.head())
print(acc_2021_cleaned['ปีที่เกิดเหตุ'].isnull().values.any())
```

```python
province_2021_count = acc_2021_cleaned['จังหวัด'].value_counts()
province_2021_count_df = pd.DataFrame({'province':province_2021_count.index,'count':province_2021_count.values,'year':2021})
print(province_2021_count_df.head(10))
```

```python
acc_2022 = pd.read_csv('accident2022.csv')
#print(acc_2020.head())

acc_2022_cleaned = acc_2022.dropna()
print(acc_2022_cleaned.head())
print(acc_2022_cleaned['ปีที่เกิดเหตุ'].isnull().values.any())
```

```python
province_2022_count = acc_2022_cleaned['จังหวัด'].value_counts()
province_2022_count_df = pd.DataFrame({'province':province_2022_count.index,'count':province_2022_count.values,'year':2022})
print(province_2022_count_df.head(5))
```

```python
combined_2020_2022_province_df = pd.concat([province_2020_count_df,province_2021_count_df,province_2022_count_df])
print(combined_2020_2022_province_df)
```

```python
import numpy as np
summed_province_each_year = combined_2020_2022_province_df.groupby(['province','year'])['count'].sum().reset_index()
#print(summed_province_each_year)

pivoted_df = summed_province_each_year.pivot(index='province',columns='year',values='count').sort_values(by=2020,ascending=False).head(5).reset_index()
#print(pivoted_df.head)
pivoted_df.index.name = 'Index'
print(pivoted_df)

ax = pivoted_df.plot(kind='bar',x='province',rot=0, width=0.7)
thai_labels = pivoted_df.head(5).province.unique().tolist()
#print(thai_labels)
eng_province = ['Bangkok','Chonburi','Chiang Mai', 'Suphan Buri', 'Nakhon Ratchasima']


ax.set_xticklabels(thai_labels, fontfamily='Tahoma', fontsize=12)
ax.set_xticklabels(eng_province, fontfamily='Tahoma', fontsize=8)

ax.set_title('Top 5 Provinces with highest car accident count by Year 2020 to 2022 ', fontweight='bold', fontsize=16)
ax.set_xlabel('Province', fontweight='bold', fontsize=14)
ax.set_ylabel('Accident Count',fontweight='bold',fontsize=14)

plt.show()

```

```python
summed_2020_2022_province_df = combined_2020_2022_province_df.groupby('province')['count'].sum().reset_index()

sorted_summed_2020_2022_province_df = summed_2020_2022_province_df.sort_values(by='count',ascending=False)

#print(summed_2020_2022_province_df)
print(sorted_summed_2020_2022_province_df)
```
