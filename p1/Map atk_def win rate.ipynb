```python
import matplotlib.pyplot as plt
import pandas as pd

# Create the first dataframe with map names and win rates
data1 = {'Map': ['Map 1', 'Map 2', 'Map 3', 'Map 4', 'Map 5', 'Map 6'],
         'Win Rate': [0.75, 0.65, 0.55, 0.85, 0.60, 0.70]}
df1 = pd.DataFrame(data1)

# Create the second dataframe with map names and win rates
data2 = {'Map': ['Map 1', 'Map 2', 'Map 3', 'Map 4', 'Map 5', 'Map 6'],
         'Win Rate': [0.70, 0.60, 0.50, 0.80, 0.55, 0.65]}
df2 = pd.DataFrame(data2)

# Plot the first set of maps
bar1 = plt.bar(df1['Map'], df1['Win Rate'], color='blue')

# Shift the x-coordinates for the second set of maps
x_coordinates = [x + 0.2 for x in range(len(df2['Map']))]

# Plot the second set of maps
bar2 = plt.bar(x_coordinates, df2['Win Rate'], color='red')

# Add a title to the graph
plt.title('Win Rate by Map')

# Label the x-axis
plt.xlabel('Map')

# Label the y-axis
plt.ylabel('Win Rate')

# Add a legend to the graph
plt.legend((bar1, bar2), ('Set 1', 'Set 2'))

# Show the graph
plt.show()
```
