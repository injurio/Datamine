```python
import requests
from urllib.parse import urlparse
from urllib.parse import parse_qs
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Scrape the data from the website

url = "https://www.valorbuff.com/weapons8"
params ={'mode':2,'rank':8}

response = requests.get(url,params=params)

soup = BeautifulSoup(response.text, "html.parser")
print(response)

```

```python
# Extract the relevant information
data = []
table = soup.find("table", {"id": "weapons"})
rows = table.find("tbody").find_all("tr")
for row in rows:
    cols = row.find_all("td")
    weapon = cols[0].text.strip()
    kills = cols[3].text.strip()
    data.append([weapon, kills])

# Store the data in a pandas DataFrame
df = pd.DataFrame(data, columns=["Weapon", "Kills"])
print(df)

```

```python
# Replace non-integer values in the Kills column with NaN
df['Kills'].replace('[^0-9]+', '', regex=True, inplace=True)

# Remove rows containing NaN
df.dropna(subset=['Kills'], inplace=True)

# Convert the Kills column to a numerical type
df["Kills"] = df["Kills"].astype(int)
df['Kills'] = df['Kills']/df['Kills'].sum() * 100

# Sort the DataFrame by the number of kills
df.sort_values("Kills", ascending=False, inplace=True)

# Get the top 5 weapons by total kill
top_5 = df.head(5)

# Plot the data as a bar graph
top_5.plot(x='Weapon', y='Kills', kind="bar", legend=False)

# Add a title and labels
plt.title("Top 5 Weapons by Total Kill")
plt.xlabel("Weapon")
plt.ylabel("Total Kill")

# Show the plot
plt.show()
```
